# Problem Definition

Crime in South Africa remains one of the most pressing social and economic challenges. High levels of violent and property-related crimes negatively impact public safety, business confidence, urban development, and the nation’s global reputation. While the South African Police Service (SAPS) publishes quarterly and annual crime statistics, these raw figures alone are often difficult to interpret, making it challenging for decision-makers to plan effective interventions.

This project aims to leverage **machine learning, programming, and data visualization** to provide actionable intelligence from available crime and contextual data. Specifically, it addresses two key objectives:

1. **Classification of Crime Hotspots**
   - Certain areas, such as neighborhoods or specific time periods (e.g., late-night weekends), experience concentrated criminal activity, often referred to as “hotspots.”
   - The task is to **predict whether a precinct or geographical/time unit qualifies as a hotspot**, using historical crime data, police staffing, geographical information, and optional socio-economic indicators.
   - This enables law enforcement and municipal authorities to **allocate patrols, community policing resources, and technology deployments more effectively**.

2. **Forecasting Crime Trends**
   - Beyond identifying current hotspots, decision-makers need to **anticipate future crime patterns**.
   - The task is to **forecast crime trends for a selected crime category and geographic unit**, revealing seasonal spikes, emerging trends, or declines due to policy changes.
   - Forecasts will cover **12–24 months** and provide **confidence intervals**, allowing authorities to proactively **adjust resource allocation and intervention strategies**.

### Key Questions Addressed
- Which precincts are most likely to experience high crime (hotspots)?  
- How will specific crime categories evolve over time across different regions?  
- How can police and municipal resources be optimally allocated to reduce crime and enhance public safety?

- Geographic focus: South African provinces, districts, and police stations.  
- Temporal coverage: 2011–2023 (or available years from datasets).

````python
import pandas as pd

# -----------------------
# Base path
data_path = "/content/drive/MyDrive/TPFINAL/"

# -----------------------
# Load datasets
crime_data = pd.read_csv(data_path + "crime_data/crime_incidents_by_category.csv")
police_coords = pd.read_csv(data_path + "geo_export_b681af05_ce59_42e8_87e6_41af282b0a6e (1).csv")
police_employed = pd.read_csv(data_path + "data.csv")
socio_econ_sa = pd.read_csv(data_path + "WDICSV.csv")

# -----------------------
# Filter socio-economic for South Africa and melt years
socio_econ_sa = socio_econ_sa[socio_econ_sa['Country Name'] == 'South Africa']
year_columns = [str(y) for y in range(1960, 2024)]
socio_econ_long = socio_econ_sa.melt(
    id_vars=['Country Name','Country Code','Indicator Name','Indicator Code'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Value'
)
socio_econ_long['Year'] = pd.to_numeric(socio_econ_long['Year'], errors='coerce')
socio_econ_long['Value'] = pd.to_numeric(socio_econ_long['Value'], errors='coerce')
socio_econ_long = socio_econ_long.dropna(subset=['Year','Value'])

# -----------------------
# Rename columns for consistency
crime_data = crime_data.rename(columns={
    'Geography': 'Province',
    'Crime Category': 'Crime Type',
    'Financial Year': 'Year',
    'Count': 'Crime Count'
})
police_coords = police_coords.rename(columns={
    'compnt_nm': 'Province',
    'location_x': 'Longitude',
    'location_y': 'Latitude'
})
police_employed = police_employed.rename(columns={
    'financial_year': 'Year',
    'number_of_employees': 'Number of Officers'
})

# -----------------------
# Merge for classification dataset (hotspot prediction)
# 1. Crime + Police coordinates
crime_coords = pd.merge(
    crime_data,
    police_coords[['Province','Latitude','Longitude']],
    on='Province',
    how='left'
)

# 2. Add police employed info (Year level)
crime_police = pd.merge(
    crime_coords,
    police_employed[['Year','Number of Officers']],
    on='Year',
    how='left'
)

# 3. Merge selected socio-economic indicators
# Filter only the indicators of interest
indicators = ['Population','Unemployment','Poverty headcount','School enrollment']
socio_filtered = socio_econ_long[socio_econ_long['Indicator Name'].isin(indicators)]
# Pivot to have indicators as columns
socio_pivot = socio_filtered.pivot(index='Year', columns='Indicator Name', values='Value').reset_index()

# 4. Merge with crime + police
crime_classification_df = pd.merge(
    crime_police,
    socio_pivot,
    on='Year',
    how='left'
)

# -----------------------
# Merge for forecasting dataset (time series)
# Aggregate crime counts by Year and Crime Type
crime_forecast = crime_data.groupby(['Year','Crime Type'])['Crime Count'].sum().reset_index()

# Merge with socio-economic indicators (pivoted)
crime_forecast_df = pd.merge(
    crime_forecast,
    socio_pivot,
    on='Year',
    how='left'
)

# -----------------------
# Preview results
print("Classification dataset:")
print(crime_classification_df.head())
print("\nForecasting dataset:")
print(crime_forecast_df.head())
````
